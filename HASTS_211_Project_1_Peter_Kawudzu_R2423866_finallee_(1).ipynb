```python
# Define parameters
ticker = "AAPL"
start_date = "2018-01-01"
end_date = "2025-12-31"

# Download data from Yahoo Finance
data = yf.download(ticker, start=start_date, end=end_date)

# Display the first few rows
display(data.head())
```

<div style="border-left:5px solid #F7941D; background:#0B0C2A; border-radius:0 6px 6px 0; padding:10px 20px; margin:18px 0 6px 0;"><span style="font-size:1.1em; font-weight:700; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif; letter-spacing:1px;">Feature Engineering</span></div>

```python
df = data.copy()

# ── Returns ──────────────────────────────────────────────────────────────────
df['Returns']     = df['Close'].pct_change()
df['Log_Returns'] = np.log(df['Close'] / df['Close'].shift(1))

# ── Price features (will demonstrate multicollinearity) ──────────────────────
df['MA_5']   = df['Close'].rolling(5).mean()
df['MA_20']  = df['Close'].rolling(20).mean()
df['MA_50']  = df['Close'].rolling(50).mean()
df['MA_200'] = df['Close'].rolling(200).mean()

# ── Volatility features ───────────────────────────────────────────────────────
df['Vol_5']  = df['Log_Returns'].rolling(5).std()  * np.sqrt(252)
df['Vol_20'] = df['Log_Returns'].rolling(20).std() * np.sqrt(252)
df['Vol_60'] = df['Log_Returns'].rolling(60).std() * np.sqrt(252)

# ── Other features ────────────────────────────────────────────────────────────
df['Momentum'] = df['Close'] / df['Close'].shift(20) - 1
df['Range']    = df['High'] - df['Low']

df = df.reset_index().dropna().reset_index(drop=True)

print(f"Clean dataset: {df.shape[0]} trading days")
display(df[['Date','Open','High','Low','Close','Volume',
            'Returns','Log_Returns','MA_5','MA_20','Vol_20']].head())
```

<div style="background:linear-gradient(90deg,#F7941D,#ffb347,#F7941D); border-radius:8px; padding:14px 24px; margin:24px 0 8px 0; text-align:center;"><span style="font-size:1.35em; font-weight:800; color:#0B0C2A; letter-spacing:2px; text-transform:uppercase; font-family:'Segoe UI',Arial,sans-serif;">CHALLENGE 1 — MULTICOLLINEARITY</span></div>

<div style="border-left:5px solid #F7941D; background:#0B0C2A; border-radius:0 6px 6px 0; padding:10px 20px; margin:18px 0 6px 0;"><span style="font-size:1.1em; font-weight:700; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif; letter-spacing:1px;">1.1 Definition</span></div>

**Multicollinearity** occurs in multiple regression when two or more predictor variables are highly linearly correlated. Given the model:

$$Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \cdots + \beta_k X_k + \varepsilon$$

**Perfect multicollinearity** exists when:
$$c_1 X_1 + c_2 X_2 + \cdots + c_k X_k = 0 \quad \text{for constants } c_i \text{ not all zero}$$

The OLS estimator $\hat{\boldsymbol{\beta}} = (\mathbf{X}^\top \mathbf{X})^{-1} \mathbf{X}^\top \mathbf{y}$ is **undefined** when $\mathbf{X}^\top \mathbf{X}$ is singular. The **Variance Inflation Factor** quantifies the problem:

$$\text{VIF}_j = \frac{1}{1 - R_j^2}$$

where $R_j^2$ is the $R^2$ from regressing $X_j$ on all other predictors. Coefficient variance inflates as:

$$\text{Var}(\hat{\beta}_j) = \frac{\sigma^2}{\text{SST}_j} \cdot \text{VIF}_j$$

<div style="border-left:5px solid #F7941D; background:#0B0C2A; border-radius:0 6px 6px 0; padding:10px 20px; margin:18px 0 6px 0;"><span style="font-size:1.1em; font-weight:700; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif; letter-spacing:1px;">1.2 Description</span></div>

Multicollinearity arises naturally in financial modeling when technical indicators—moving averages, momentum signals, and volatility measures—are constructed from the same underlying price series over different look-back windows. Although individually meaningful, these features can be so highly correlated that the regression model cannot isolate each predictor's independent contribution, rendering individual coefficient estimates unreliable even while the overall model fit (R²) appears strong.

<div style="border-left:5px solid #F7941D; background:#0B0C2A; border-radius:0 6px 6px 0; padding:10px 20px; margin:18px 0 6px 0;"><span style="font-size:1.1em; font-weight:700; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif; letter-spacing:1px;">1.3 Demonstration — Numerical Example</span></div>

<div style="border-left:5px solid #F7941D; background:#0B0C2A; border-radius:0 6px 6px 0; padding:10px 20px; margin:18px 0 6px 0;"><span style="font-size:1.1em; font-weight:700; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif; letter-spacing:1px;">2.1 Definition</span></div>

**Skewness** is the third standardized moment, measuring asymmetry around the mean:

$$\text{Skewness} = \gamma_1 = \mathbb{E}\left[\left(\frac{X - \mu}{\sigma}\right)^3\right] = \frac{\mu_3}{\sigma^3}$$

Sample estimator:

$$\hat{\gamma}_1 = \frac{\frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})^3}{\left(\frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})^2\right)^{3/2}}$$

Interpretation:
- $\gamma_1 = 0$ → symmetric (Normal)
- $\gamma_1 < 0$ → **negative (left) skew** — long left tail, large losses more probable  
- $\gamma_1 > 0$ → **positive (right) skew** — long right tail

<div style="border-left:5px solid #F7941D; background:#0B0C2A; border-radius:0 6px 6px 0; padding:10px 20px; margin:18px 0 6px 0;"><span style="font-size:1.1em; font-weight:700; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif; letter-spacing:1px;">2.2 Description</span></div>

Skewness in financial returns reflects the well-documented empirical reality that large negative moves occur more frequently and with greater magnitude than large positive moves—called "left-tail risk" or "crash risk." When volatility models incorrectly assume symmetric normal returns, they systematically misprice options (particularly puts), understate tail risk, and produce VaR estimates that underestimate the probability of large losses.

<div style="border-left:5px solid #F7941D; background:#0B0C2A; border-radius:0 6px 6px 0; padding:10px 20px; margin:18px 0 6px 0;"><span style="font-size:1.1em; font-weight:700; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif; letter-spacing:1px;">2.3 Demonstration — Numerical Example</span></div>

<div style="background:linear-gradient(90deg,#F7941D,#ffb347,#F7941D); border-radius:8px; padding:14px 24px; margin:24px 0 8px 0; text-align:center;"><span style="font-size:1.35em; font-weight:800; color:#0B0C2A; letter-spacing:2px; text-transform:uppercase; font-family:'Segoe UI',Arial,sans-serif;">CHALLENGE 3 — SENSITIVITY TO OUTLIERS</span></div>

<div style="border-left:5px solid #F7941D; background:#0B0C2A; border-radius:0 6px 6px 0; padding:10px 20px; margin:18px 0 6px 0;"><span style="font-size:1.1em; font-weight:700; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif; letter-spacing:1px;">3.1 Definition</span></div>

**Sensitivity to outliers** is the degree to which statistical estimators are distorted by extreme observations. The **breakdown point** $\varepsilon^*$ measures robustness:

$$\varepsilon^*(T, \mathbf{x}) = \min\left\{\frac{m}{n}: \sup_{\mathbf{x}'_m} \|T(\mathbf{x}'_m) - T(\mathbf{x})\| = \infty\right\}$$

The OLS estimator has breakdown point $\varepsilon^* = 1/n \approx 0$. Leverage and influence are measured by:

$$h_{ii} = \mathbf{x}_i^\top(\mathbf{X}^\top\mathbf{X})^{-1}\mathbf{x}_i \quad \text{(hat matrix diagonal)}$$

$$\text{Cook's Distance} = D_i = \frac{(\hat{\boldsymbol{\beta}} - \hat{\boldsymbol{\beta}}_{(-i)})^\top (\mathbf{X}^\top\mathbf{X})(\hat{\boldsymbol{\beta}} - \hat{\boldsymbol{\beta}}_{(-i)})}{p \cdot \hat{\sigma}^2}$$

<div style="border-left:5px solid #F7941D; background:#0B0C2A; border-radius:0 6px 6px 0; padding:10px 20px; margin:18px 0 6px 0;"><span style="font-size:1.1em; font-weight:700; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif; letter-spacing:1px;">3.2 Description</span></div>

Financial time series routinely experience shock events—earnings surprises, regulatory announcements, or market panics—that generate returns orders of magnitude beyond what standard models expect. Because classical OLS minimizes the sum of squared residuals, these extreme observations receive disproportionate weight, pulling estimated coefficients and volatility forecasts toward the extreme event rather than representing typical market behavior.

<div style="border-left:5px solid #F7941D; background:#0B0C2A; border-radius:0 6px 6px 0; padding:10px 20px; margin:18px 0 6px 0;"><span style="font-size:1.1em; font-weight:700; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif; letter-spacing:1px;">3.3 Demonstration — Numerical Example</span></div>

<div style="background:linear-gradient(90deg,#F7941D,#ffb347,#F7941D); border-radius:8px; padding:14px 24px; margin:24px 0 8px 0; text-align:center;"><span style="font-size:1.35em; font-weight:800; color:#0B0C2A; letter-spacing:2px; text-transform:uppercase; font-family:'Segoe UI',Arial,sans-serif;">CHALLENGE 4 — OVERFITTING</span></div>

<div style="border-left:5px solid #F7941D; background:#0B0C2A; border-radius:0 6px 6px 0; padding:10px 20px; margin:18px 0 6px 0;"><span style="font-size:1.1em; font-weight:700; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif; letter-spacing:1px;">4.1 Definition</span></div>

**Overfitting** occurs when a model learns noise in the training sample rather than the true data-generating process. The expected prediction error decomposes as the **Bias-Variance Tradeoff**:

$$\text{EPE}(\hat{f}) = \underbrace{\text{Bias}^2[\hat{f}(x)]}_{\text{systematic error}} + \underbrace{\text{Var}[\hat{f}(x)]}_{\text{estimation noise}} + \underbrace{\sigma^2_\varepsilon}_{\text{irreducible}}$$

An overfitted model achieves near-zero training error by minimising bias at the cost of extremely high variance:

$$\text{Train MSE} \ll \text{Test MSE}$$

Model complexity is penalised by information criteria. **AIC** and **BIC**:

$$\text{AIC} = -2\ell(\hat{\theta}) + 2k \qquad \text{BIC} = -2\ell(\hat{\theta}) + k\ln(n)$$

where $\ell$ is the log-likelihood and $k$ the number of free parameters.

<div style="border-left:5px solid #F7941D; background:#0B0C2A; border-radius:0 6px 6px 0; padding:10px 20px; margin:18px 0 6px 0;"><span style="font-size:1.1em; font-weight:700; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif; letter-spacing:1px;">4.2 Description</span></div>

In quantitative finance, overfitting is the primary cause of strategy failure at live deployment: a volatility model that looks highly predictive in-sample may have merely memorised idiosyncratic features of a historical regime, performing no better than chance—or catastrophically worse—when deployed in production. This is especially acute for derivatives pricing, where a model calibrated too tightly to yesterday's implied volatility surface may produce arbitrage-prone prices the next day.

<div style="border-left:5px solid #F7941D; background:#0B0C2A; border-radius:0 6px 6px 0; padding:10px 20px; margin:18px 0 6px 0;"><span style="font-size:1.1em; font-weight:700; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif; letter-spacing:1px;">4.3 Demonstration — Numerical Example</span></div>

```python
# ── AIC / BIC comparison ─────────────────────────────────
# Re-initialize X_raw and y_raw for this cell's context to ensure consistency
X_raw = df['Vol_20'].values.reshape(-1, 1)
y_raw = df['Vol_20'].shift(-1).dropna().values
X_raw = X_raw[:len(y_raw)] # Ensure X_raw and y_raw have the same length

n_tr = len(X_train)
ic_records = []
for d in [1, 2, 3, 5, 8]:
    pipe = Pipeline([('poly', PolynomialFeatures(degree=d)),
                     ('sc',   StandardScaler()),
                     ('m',    LinearRegression())])
    pipe.fit(X_train, y_train)
    resid = y_train - pipe.predict(X_train)
    sse   = (resid**2).sum()
    k     = d + 2
    ll    = -n_tr/2 * np.log(2*np.pi*sse/n_tr) - n_tr/2
    ic_records.append({'Degree': d, 'Params k': k,
                       'AIC': -2*ll + 2*k,
                       'BIC': -2*ll + k*np.log(n_tr)})

ic_df = pd.DataFrame(ic_records)

d12 = res_df[res_df['Degree']==12].iloc[0]

# ── Walk-forward OOS R² (30-day ahead) ────────────────

train_initial_size = int(len(X_raw) * 0.7)
forecast_horizon = 30

wf_simple = []
wf_complex = []

for i in range(train_initial_size, len(X_raw) - forecast_horizon):
    X_train_wf, y_train_wf = X_raw[:i], y_raw[:i]
    X_test_wf, y_test_wf = X_raw[i:i + forecast_horizon], y_raw[i:i + forecast_horizon]

    # Simple model (Degree 1)
    pipe_simple = Pipeline([
        ('poly',   PolynomialFeatures(degree=1)),
        ('scale',  StandardScaler()),
        ('model',  LinearRegression())
    ])
    pipe_simple.fit(X_train_wf, y_train_wf)
    y_pred_simple = pipe_simple.predict(X_test_wf)
    wf_simple.append(r2_score(y_test_wf, y_pred_simple))

    # Complex model (Degree 12)
    pipe_complex = Pipeline([
        ('poly',   PolynomialFeatures(degree=12)),
        ('scale',  StandardScaler()),
        ('model',  LinearRegression())
    ])
    pipe_complex.fit(X_train_wf, y_train_wf)
    y_pred_complex = pipe_complex.predict(X_test_wf)
    wf_complex.append(r2_score(y_test_wf, y_pred_complex))


print("DIAGNOSIS ─ Detecting Overfitting")
print("=" * 58)
print(f"\n1. Train-Test R² Gap (Degree 12 polynomial):")
print(f"   Train R² = {d12['Train R²']:.4f}  |  Test R² = {d12['Test R²']:.6f}")
print(f"   Gap      = {d12['Gap']:.4f}  →  {'OVERFIT' if d12['Gap'] > 0.1 else 'OK'}")

print(f"\n2. Information Criteria (lower = better):")
display(ic_df.set_index('Degree').style.format({'AIC':'{:.2f}','BIC':'{:.2f}'}))

best_aic = ic_df.loc[ic_df['AIC'].idxmin(), 'Degree']
best_bic = ic_df.loc[ic_df['BIC'].idxmin(), 'Degree']
print(f"   Best model by AIC : Degree {best_aic}")
print(f"   Best model by BIC : Degree {best_bic}")

print(f"\n3. Walk-forward OOS R² (30-day ahead):")
print(f"   Mean R² — Degree 1  : {np.mean(wf_simple):.4f}")
print(f"   Mean R² — Degree 12 : {np.mean(wf_complex):.4f}")
print(f"   Complex model generalises {'WORSE' if np.mean(wf_complex) < np.mean(wf_simple) else 'BETTER'}")
```

**What the results show:**  
More complex models appear to explain historical volatility extremely well—until tested on new data, where their performance collapses. A high-complexity polynomial model achieves near-perfect accuracy on training data but performs no better than chance on out-of-sample periods. A simple, parsimonious GARCH(1,1) model consistently outperforms complex alternatives when tested on future data.

**Recommended course of action:**  
Always evaluate volatility models on data they have never seen before (out-of-sample testing). Prefer parsimonious models with few parameters—the industry-standard GARCH(1,1) with 3 parameters is robust across market regimes and is far less prone to overfitting than elaborate multi-factor models. When additional complexity is considered, apply formal model-selection criteria (AIC/BIC) to confirm that added parameters genuinely improve generalization, not just in-sample fit.

**Key factors driving this issue:**  
The principal risk factors for overfitting on the derivatives desk are: (1) calibrating models to short historical windows that may not be representative; (2) adding features or model parameters without formal penalization; and (3) backtesting on the same data used for model selection. A strict separation between calibration data and validation data is the single most important operational safeguard.

<div style="border-left:5px solid #F7941D; background:#0B0C2A; border-radius:0 6px 6px 0; padding:10px 20px; margin:18px 0 6px 0;"><span style="font-size:1.1em; font-weight:700; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif; letter-spacing:1px;">Identification of the Underlying Challenge</span></div>

Across all four sections of this handbook, the challenges of multicollinearity, skewness, outlier sensitivity, and overfitting each stem from a common, deeper problem: **how to correctly prioritize and weight the factors that drive volatility**. This is the **"prioritizing factors"** challenge.

<div style="border-bottom:1px solid rgba(247,148,29,0.3); padding:6px 4px 6px 12px; margin:14px 0 4px 0;"><span style="font-size:1em; font-weight:600; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif;">Formal Statement</span></div>

Factor prioritization is the problem of determining which explanatory variables most meaningfully and reliably predict the target—AAPL's volatility—and assigning those variables appropriate model weight. Given a feature set $\{X_1, X_2, \ldots, X_p\}$, the goal is:

$$\mathcal{S}^*, \boldsymbol{\beta}^* = \arg\min_{\mathcal{S}, \boldsymbol{\beta}} \, \mathbb{E}_{\text{new data}} \left[ \left(Y - \sum_{j \in \mathcal{S}} \beta_j X_j \right)^2 \right]$$

<div style="border-bottom:1px solid rgba(247,148,29,0.3); padding:6px 4px 6px 12px; margin:14px 0 4px 0;"><span style="font-size:1em; font-weight:600; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif;">How Each Challenge Maps to Factor Prioritization</span></div>

| Section | Challenge | How it distorts factor weights | Our solution |
|---------|-----------|-------------------------------|-------------|
| 1 | Multicollinearity | Cannot isolate which signal matters; SE inflated | Lasso/Ridge/PCA |
| 2 | Skewness | Normal models under-weight left-tail observations | Skew-t / GJR-GARCH |
| 3 | Outlier Sensitivity | Extreme days dominate coefficient estimation | MAD / Winsorization |
| 4 | Overfitting | Model weights noise, ignores signal | Walk-forward / AIC/BIC |

```python
# ── STEP 4 — DIRECTIONS: Permutation Importance & Forward Stepwise Selection ──
from sklearn.ensemble import RandomForestRegressor
from sklearn.inspection import permutation_importance
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from scipy.stats import spearmanr

# ── Features available in df ──────────────────────────────────────────────────
feature_cols_p4 = ['MA_5','MA_20','MA_50','MA_200','Vol_5','Vol_20','Vol_60','Momentum']

# ── Build working dataset (avoid duplicate Vol_5 column) ─────────────────────
df_p4 = df[feature_cols_p4].dropna().copy()
df_p4['Target'] = df['Vol_5'].shift(-5).loc[df_p4.index]   # 5-day ahead vol
df_p4 = df_p4.dropna().reset_index(drop=True)

X_prio = df_p4[feature_cols_p4].values
y_prio = df_p4['Target'].values

# ── 80/20 time-based split ────────────────────────────────────────────────────
split  = int(0.80 * len(X_prio))
X_tr, X_te = X_prio[:split], X_prio[split:]
y_tr, y_te = y_prio[:split], y_prio[split:]

# ── Random Forest permutation importance ─────────────────────────────────────
rf = RandomForestRegressor(n_estimators=200, max_depth=5, random_state=42)
rf.fit(X_tr, y_tr)
perm = permutation_importance(rf, X_te, y_te, n_repeats=15, random_state=42)

imp_df = pd.DataFrame({
    'Feature'   : feature_cols_p4,
    'Importance': perm.importances_mean,
    'Std'       : perm.importances_std,
}).sort_values('Importance', ascending=False).reset_index(drop=True)

# ── Forward stepwise (OOS R2) ─────────────────────────────────────────────────
selected_fs, r2_steps = [], []
for _ in range(len(feature_cols_p4)):
    best_r2_s, best_f = -np.inf, None
    for f in [x for x in feature_cols_p4 if x not in selected_fs]:
        idx = [feature_cols_p4.index(c) for c in selected_fs + [f]]
        m   = LinearRegression().fit(X_tr[:, idx], y_tr)
        r2v = r2_score(y_te, m.predict(X_te[:, idx]))
        if r2v > best_r2_s:
            best_r2_s, best_f = r2v, f
    selected_fs.append(best_f)
    r2_steps.append(best_r2_s)

# ── Spearman correlations ──────────────────────────────────────────────────────
spearman = {f: abs(spearmanr(df_p4[f], df_p4['Target'])[0]) for f in feature_cols_p4}

# ── Print results ─────────────────────────────────────────────────────────────
print("STEP 4 — FACTOR PRIORITIZATION: Resolution")
print("=" * 62)
print("\n1. Permutation Importance (Random Forest, OOS):")
max_imp = max(imp_df['Importance'].max(), 1e-9)
for _, row in imp_df.iterrows():
    bar = chr(9608) * max(1, int(row['Importance']/max_imp * 20))
    print(f"   {row['Feature']:12s}: {bar:<22} {row['Importance']:.5f} +/- {row['Std']:.5f}")

best_k = int(np.argmax(r2_steps))
print("\n2. Forward Stepwise Selection (OOS R2):")
for i, (f, r2v) in enumerate(zip(selected_fs, r2_steps)):
    marker = ' <- optimal' if i == best_k else ''
    print(f"   Step {i+1}: Add {f:12s}  ->  R2 = {r2v:.4f}{marker}")

print("\n3. Spearman |rho| with Target (5-day Ahead Vol):")
for f, rho in sorted(spearman.items(), key=lambda x: -x[1]):
    print(f"   {f:12s}: |rho| = {rho:.4f}")

top3 = ', '.join(imp_df['Feature'].head(3).tolist())
weak = min(spearman, key=spearman.get)
print(f"\nTop 3 factors (permutation importance): {top3}")
print(f"Optimal factor count (stepwise BIC equivalent): {best_k + 1}")
print(f"Weakest factor (lowest univariate relevance): {weak}")
```

```python
# ── STEP 4 — DIAGRAM: Factor Prioritization Dashboard ────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('STEP 4 — Prioritizing Factors: Resolution Framework (AAPL 2018-2025)',
             fontsize=13, fontweight='bold')

# Panel A: Permutation Importance
colors_p = [GREEN if r > 0 else RED for r in imp_df['Importance']]
axes[0].barh(imp_df['Feature'], imp_df['Importance'],
             xerr=imp_df['Std'], color=colors_p, edgecolor='white')
axes[0].set_xlabel('Permutation Importance (drop in OOS R2)')
axes[0].set_title('Panel A: Feature Importance\n(Random Forest, Out-of-Sample)', fontweight='bold')
axes[0].invert_yaxis()

# Panel B: Stepwise R2
axes[1].bar(range(1, len(selected_fs)+1), r2_steps, color=BLUE, alpha=0.8, edgecolor='white')
axes[1].plot(range(1, len(selected_fs)+1), r2_steps, 'ro-', ms=6, label='OOS R2')
best_k_disp = int(np.argmax(r2_steps)) + 1
axes[1].axvline(best_k_disp, color='gold', ls='--', lw=2, label=f'Optimal: {best_k_disp} factors')
axes[1].set_xticks(range(1, len(selected_fs)+1))
axes[1].set_xticklabels([f'{i}\n({selected_fs[i-1][:5]})' for i in range(1,len(selected_fs)+1)],
                         fontsize=7)
axes[1].set_xlabel('Number of Factors (Forward Stepwise)')
axes[1].set_ylabel('OOS R2')
axes[1].set_title('Panel B: Stepwise Factor Addition\n(OOS R2 by step)', fontweight='bold')
axes[1].legend(fontsize=9)

# Panel C: Spearman correlation
sp_vals = pd.Series(spearman).sort_values(ascending=False)
bar_cols = [GREEN if v > 0.3 else (ORANGE if v > 0.15 else RED) for v in sp_vals]
axes[2].barh(sp_vals.index, sp_vals.values, color=bar_cols, edgecolor='white')
axes[2].axvline(0.3,  color=GREEN,  ls='--', lw=1.5, label='Strong (>0.30)')
axes[2].axvline(0.15, color=ORANGE, ls='--', lw=1.5, label='Moderate (>0.15)')
axes[2].set_xlabel('|Spearman rho| with 5-day Ahead Vol')
axes[2].set_title('Panel C: Univariate Factor Relevance\n(Spearman rank correlation)', fontweight='bold')
axes[2].legend(fontsize=8)
axes[2].invert_yaxis()

plt.tight_layout()
plt.savefig('fig_factor_prioritization.png', dpi=150, bbox_inches='tight')
plt.show()
print("Factor prioritization diagram saved.")
```
